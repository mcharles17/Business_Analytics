```python
#Load in files and concat to one dataframe
one = pd.read_csv('bloomfield.csv')
two = pd.read_csv('bellevile.csv')
three = pd.read_csv('therest.csv')
house_data = pd.concat((one, two, three), axis=0)
house_data.reset_index(inplace = True)
house_data.drop(['Unnamed: 0', 'index'], axis=1, inplace=True)
house_data
```

```python
#Clean data
house_data.drop_duplicates(inplace=True)
house_data.loc[:,'Listing Agent'] = house_data.loc[:, 'Listing Agent'].astype('string')
house_data.loc[:,'Listing Brokerage'] = house_data.loc[:, 'Listing Brokerage'].astype('string')
house_data.loc[:,"Buyer's Agent"] = house_data.loc[:, "Buyer's Agent"].astype('string')
house_data.loc[:,"Buyer's Broker"] = house_data.loc[:, "Buyer's Broker"].astype('string')
house_data.loc[:,'Listing Price'] = house_data.loc[:, 'Listing Price'].astype('string')
house_data.loc[:,'Address'] = house_data.loc[:, 'Address'].astype('string')
house_data.loc[:,'Town'] = house_data.loc[:, 'Town'].astype('string')
house_data.loc[:,'Sale Price'] = house_data.loc[:, 'Sale Price'].astype('string')
house_data.loc[:,'Terms'] = house_data.loc[:, 'Terms'].astype('string')
house_data.loc[:,'Style'] = house_data.loc[:, 'Style'].astype('string')
```

```python
def ends(x):
    return x[:-14]
house_data['Listing Agent'] = house_data['Listing Agent'].apply(ends)
house_data["Buyer's Agent"] = house_data["Buyer's Agent"].apply(ends)
def beginning(x):
    return x[10:]
house_data['Listing Agent'] = house_data['Listing Agent'].apply(beginning)
house_data["Buyer's Agent"] = house_data["Buyer's Agent"].apply(beginning)
```

```python
def ends(x):
    index = x.find('(')
    return x[:index-1]
house_data['Listing Brokerage'] = house_data['Listing Brokerage'].apply(ends)
house_data["Buyer's Broker"] = house_data["Buyer's Broker"].apply(ends)
def beginning(x):
    return x[9:]
house_data['Listing Brokerage'] = house_data['Listing Brokerage'].apply(beginning)
house_data["Buyer's Broker"] = house_data["Buyer's Broker"].apply(beginning)
```

```python
def money(x):
    return x[1:]
house_data['Listing Price'] = house_data['Listing Price'].apply(money)
house_data['Sale Price'] = house_data['Sale Price'].apply(money)
```

```python
house_data.Address = house_data.Address.apply(lambda x: x[2:])
house_data.Address = house_data.Address.apply(lambda x: x[:-4])
```

```python
house_data.Town = house_data.Town.apply(lambda x: x[:-4])
```
